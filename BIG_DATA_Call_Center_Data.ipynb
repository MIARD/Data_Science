##Convert date_time column from string to datetime

```python
df_pd = df.toPandas()
type(df_pd)
```

```python
import pandas as pd
dtf = df_pd[['date_time']].copy()
dtf['date_time'] = pd.to_datetime(dtf['date_time'])
```

```python
dtf['Hour'] = dtf['date_time'].dt.hour
dtf['Month'] = dtf['date_time'].dt.month_name()
dtf['Day'] = dtf['date_time'].dt.day_name()
dtf['Date'] = dtf['date_time'].dt.date
dtf['Year'] = dtf['date_time'].dt.year
```

```python
dtf['talk_sec'] = df_pd['talk_sec']
dtf['talk_min'] = df_pd['talk_sec'] / 60
```

```python
#Create DataFrame by changing schema
sdf = spark.createDataFrame(dtf)
sdf.printSchema()
sdf = sdf.withColumn('Month_n', month(sdf.date_time))
```

```python
sdf = sdf.withColumn('Day_n', dayofweek(sdf.date_time))
sdf.printSchema()
```

```python
sdf.createOrReplaceTempView('callCenter')
query = spark.sql("select Day_n from callCenter")
query.show()
```

```python
query = spark.sql('''
Select
Hour, count(talk_sec) as total_calls
, sum(talk_sec) as total_call_sec
, avg(talk_sec) as avg_talk_sec
, min(talk_sec) as min_talk_sec
, max(talk_sec) as max_talk_sec
from callCenter
group by Hour order by Hour
''')
query.show()
```

```python
query = spark.sql('''
Select
  Day,  count(talk_sec) as total_calls
, sum(talk_sec) as total_call_sec
, avg(talk_sec) as avg_talk_sec
, min(talk_sec) as min_talk_sec
, max(talk_sec) as max_talk_sec
from callCenter
group by Day order by MAX(Day_n)
''')
query.show()
```

## Monthly Data of Year 2021

```python
query = spark.sql('''
Select
Month, count(talk_sec) as total_calls
, sum(talk_sec) as total_call_sec
, round(avg(talk_sec)) as avg_talk_sec
, min(talk_sec) as min_talk_sec
, max(talk_sec) as max_talk_sec
from callCenter
group by Month order by MAX(Month_n)
''')
query.show()
```

```python
query = spark.sql('''
Select
Year, count(talk_sec) as total_calls
, sum(talk_sec) as total_call_sec
, round(avg(talk_sec)) as avg_talk_sec
, min(talk_sec) as min_talk_sec
, max(talk_sec) as max_talk_sec
from callCenter
group by Year
''')
query.show()
```
